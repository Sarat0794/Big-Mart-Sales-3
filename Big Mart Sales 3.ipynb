```python
combined_dataset['Item_Fat_Content']=combined_dataset['Item_Fat_Content'].replace(['LF','reg','low fat'],['Low Fat','Regular','Low Fat'])
```

```python
combined_dataset['Item_Fat_Content'].value_counts()
```

```python
#To convert the year to the no. of years since 2013
combined_dataset['Outlet_Establishment_Year']=combined_dataset['Outlet_Establishment_Year'].apply(lambda x:2013-x)
```

```python
combined_dataset['Outlet_Establishment_Year'].value_counts()
```
